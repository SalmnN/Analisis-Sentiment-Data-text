# **Cleaning data.**




**Removoe URLs, Remove Non-Alphabetic, Remove a href, Remove Extra Spaces, dan Case Folding**

```python
import re
import pandas as pd

# Define the clean_comment function
def clean_comment(comment):
    comment = re.sub(r'http\S+', '', comment)  # Remove URLs
    comment = re.sub(r'[^A-Za-z\s]', '', comment)  # Remove non-alphabetic characters
    comment = re.sub(r'a href', '', comment)  # Remove 'a href' tag
    comment = re.sub(r'\s+', ' ', comment)  # Remove extra spaces
    comment = comment.lower()  # Convert to lowercase
    return comment

# Apply the clean_comment function to the 'textDisplay' column
dataCleaning = dataMask.copy()
dataCleaning['clean_comment'] = dataCleaning['textDisplay'].apply(clean_comment)
dataCleaning = dataCleaning['clean_comment']
# Display the cleaned data
dfDataCleaning = pd.DataFrame(dataCleaning)
display(dfDataCleaning)
```

```python
import pandas as pd
import re

# Fungsi untuk membaca kamus dari CSV tanpa header
def baca_kamus_dari_csv(nama_file):
    df = pd.read_csv(nama_file, delimiter=';', header=None, names=['tidak_baku', 'kata_baku'])
    kamus = dict(zip(df['tidak_baku'], df['kata_baku']))
    return kamus

# Fungsi untuk mengganti kata menggunakan kamus
def ganti_kata_baku(teks, kamus):
    def pengganti(match):
        kata = match.group(0)
        return kamus.get(kata.lower(), kata)

    pola = re.compile(r'\b(' + '|'.join(re.escape(kata) for kata in kamus.keys()) + r')\b', re.IGNORECASE)
    teks_baru = pola.sub(pengganti, teks)
    return teks_baru

# Nama file CSV
nama_file_kamus = '/content/drive/MyDrive/RF/kamus_singkatan.csv'

# Membaca kamus dari CSV
kamus = baca_kamus_dari_csv(nama_file_kamus)

# Mengganti kata-kata dalam dataCleaning_XL menggunakan kamus
dataCleaning_Kamus = []
for teks in dataCleaning_XL:
    teks_baru = ganti_kata_baku(teks, kamus)
    dataCleaning_Kamus.append(teks_baru)

# Membuat DataFrame dari dataCleaning_XL dan dataCleaning_Kamus
dfDataCleaning_Kamus = pd.DataFrame({'Teks Awal': dataCleaning_XL, 'Teks Baru': dataCleaning_Kamus})

# Menampilkan DataFrame
print(dfDataCleaning_Kamus)
```

# **Stemming**

```python
import nltk
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
import pandas as pd # Import pandas

# Download NLTK data (if not already downloaded)
nltk.download('punkt')

# Create stemmer
factory = StemmerFactory()
stemmer = factory.create_stemmer()

def stem_text(text):
    words = nltk.word_tokenize(text)  # Tokenize the text
    stemmed_words = [stemmer.stem(word) for word in words]  # Stem each word
    return " ".join(stemmed_words)  # Join the stemmed words back into a string

# Convert the list to a Pandas Series
dataStopWordSeries = pd.Series(dataStopWord)

# Apply the stem_text function to each element in the Series
stemmed_series = dataStopWordSeries.apply(stem_text) # Use the Series here

# Print the first few stemmed comments (for example)
dfDataStemming = pd.DataFrame(stemmed_series)
display(dfDataStemming)
```

# **TF IDF**

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd

# Assuming `dataTokenization` is your DataFrame
data_fix = dataTokenization.copy()

# Join the tokens back into strings
# Iterate over the values in the first column (index 0) of the DataFrame
data_fix = [' '.join(tokens) for tokens in data_fix.iloc[:, 0]]

# Membuat TF-IDF Vectorizer
tfidf_vectorizer = TfidfVectorizer(max_features=1000)  # Menggunakan 1000 fitur teratas

# Mengubah teks ke dalam representasi TF-IDF
tfidf_matrix = tfidf_vectorizer.fit_transform(data_fix)

# Membuat DataFrame dari hasil TF-IDF
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf_vectorizer.get_feature_names_out())

# Menampilkan hasil
dfTfIdf = pd.DataFrame(tfidf_df)
display(dfTfIdf)
```

# **Ektraksi Fitur Analisis Sentimen**


```python
# Menginstall TextBlob jika belum terinstall
!pip install textblob
from textblob import TextBlob
import pandas as pd

# Check for and handle non-string tokens
def safe_join(tokens):
    return ' '.join([t for t in tokens if isinstance(t, str)])

# Membuat kolom untuk polaritas sentimen
dataTokenization['polarity'] = dataTokenization.apply(lambda tokens: TextBlob(safe_join(tokens)).sentiment.polarity)

# Membuat fungsi untuk mengklasifikasikan sentimen
def classify_sentiment(polarity):
    if polarity < 0:
        return 'negatif'
    elif polarity == 0:
        return 'netral'
    else:
        return 'positif'

# Mengklasifikasikan sentimen berdasarkan polaritas
dataTokenization['sentiment'] = dataTokenization['polarity'].apply(classify_sentiment)

# Memisahkan komentar berdasarkan sentimen
negatif_df = dataTokenization[dataTokenization['sentiment'] == 'negatif']
netral_df = dataTokenization[dataTokenization['sentiment'] == 'netral']
positif_df = dataTokenization[dataTokenization['sentiment'] == 'positif']

# Menghapus kolom 'polarity' agar output lebih sederhana
dataTokenization.drop(columns=['polarity'], inplace=True)

# Menampilkan DataFrame dengan pelabelan sentimen
pd.DataFrame(dataTokenization)
```
